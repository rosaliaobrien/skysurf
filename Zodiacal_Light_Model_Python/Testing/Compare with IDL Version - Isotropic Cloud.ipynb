```python
wavelength_arr = np.random.uniform(0.2,3.5,100)
day_arr = np.random.uniform(0,365,100)
lon_arr = np.random.uniform(0,360,100)
lat_arr = np.random.uniform(-90,90,100)
wavelength_arr
```

```python
ratio_list = []

for wave in wavelength_arr:

    wave = np.round(wave, 5)

    IDL.wave = wave
    IDL.day = day_arr
    IDL.lon = lon_arr
    IDL.lat = lat_arr
    
    idl_run_str = "zodi = get_zmod(wave,'skysurf',day,lon,lat,new_iso_comp=1,iso_comp_only=1)"
    IDL.run(idl_run_str)
    
    zodi_python = get_zmod(wave, 'skysurf', day_arr, lon_arr, lat_arr, new_iso_comp=True, iso_comp_only=True)
    
    ratio = IDL.zodi/zodi_python
    
    # Find indices where ratio is too high or too low
    bad_high = [i for i, r in enumerate(ratio) if r > 1.00001]
    bad_low  = [i for i, r in enumerate(ratio) if r < 0.99999]

    if bad_high or bad_low:
        print(f"\nBad wave: {wave}")

        print('\n')
        for i in bad_high:
            print('Zodi Python:', zodi_python[i])
            print('Zodi IDL:', IDL.zodi[i])
            print('Percent err:', (1-ratio[i])*100)
            print('Day:', day_arr[i])
            print('Lon:', lon_arr[i])
            print('Lat:', lat_arr[i])
            print('\n')

        print('\n')
        for i in bad_low:
            print('Zodi Python:', zodi_python[i])
            print('Zodi IDL:', IDL.zodi[i])
            print('Percent err:', (1-ratio[i])*100)
            print('Day:', day_arr[i])
            print('Lon:', lon_arr[i])
            print('Lat:', lat_arr[i])
            print('\n')

        # print('\n')
```
